```python
from langchain_ollama import ChatOllama
from openai import OpenAI
from dotenv import load_dotenv

load_dotenv()
cloud_llm = OpenAI()

local_llm = ChatOllama(model = "llama3.2:3b")
```

```python
data = """ The 13-inch MacBook Air comes with an M5 chip featuring a 10-core CPU, 10-core GPU and 16-core Neural Engine, 
            24GB unified memory, 1TB SSD storage, a 70W USB-C Power Adapter, Indian English Backlit Magic Keyboard with Touch ID, 
            two Thunderbolt 4 ports, MagSafe 3 port, 3.5mm headphone jack, Wi-Fi 7, Bluetooth 6, support for up to two external displays, 
            available in Sky Blue color"""
```

## Task 1

```python
prompt = "Extract the information from this given data"

messages = [
    {"role": "system", "content": prompt},
    {"role": "user", "content": data}
]
```

### Ollama

```python
response = local_llm.invoke(messages)

content = response.content
input_tokens  = response.usage_metadata['input_tokens']    
output_tokens = response.usage_metadata['output_tokens']   
total_tokens  = response.usage_metadata['total_tokens'] 

print(f"Content : {content}")
print(f"input_tokens : {input_tokens}")
print(f"output_tokens : {output_tokens}")
print(f"total_tokens : {total_tokens}")
```

### OpenAI

```python
response = cloud_llm.chat.completions.create(
    model="gpt-4o-mini",
    messages=messages,
)

content = response.choices[0].message.content
input_tokens  = response.usage.prompt_tokens
output_tokens = response.usage.completion_tokens
total_tokens  = response.usage.total_tokens

print(f"Content : {content}")
print(f"input_tokens : {input_tokens}")
print(f"output_tokens : {output_tokens}")
print(f"total_tokens : {total_tokens}")
```

## Task 3

```python
from pydantic import BaseModel, Field
from typing import Optional


class Processor(BaseModel):
    chip: str = Field(description="The chip name, e.g. 'M5'")
    cpu_cores: int = Field(description="Number of CPU cores")
    gpu_cores: int = Field(description="Number of GPU cores")
    neural_engine_cores: int = Field(description="Number of Neural Engine cores")


class MacBookSpec(BaseModel):
    model: str = Field(description="Product model name, e.g. '13-inch MacBook Air'")
    processor: Processor
    unified_memory: str = Field(description="Unified memory, e.g. '24GB'")
    storage: str = Field(description="SSD storage capacity, e.g. '1TB'")
    power_adapter: str = Field(description="Power adapter spec, e.g. '70W USB-C'")
    keyboard: str = Field(description="Keyboard description")
    thunderbolt_ports: int = Field(description="Number of Thunderbolt 4 ports")
    has_magsafe: bool = Field(description="Whether it has a MagSafe 3 port")
    has_headphone_jack: bool = Field(description="Whether it has a 3.5mm headphone jack")
    wifi: str = Field(description="Wi-Fi standard, e.g. 'Wi-Fi 7'")
    bluetooth: str = Field(description="Bluetooth version, e.g. 'Bluetooth 6'")
    max_external_displays: int = Field(description="Max number of external displays")
    color: str = Field(description="Available color")
```

```python
prompt = "Extract the information from this given data in the given json structure"

messages = [
    {"role": "system", "content": prompt},
    {"role": "user", "content": data}
]
```

```python
structured_local = local_llm.with_structured_output(MacBookSpec, include_raw=True)

result = structured_local.invoke(messages)

content       = result["parsed"]
input_tokens  = result["raw"].usage_metadata["input_tokens"]
output_tokens = result["raw"].usage_metadata["output_tokens"]
total_tokens  = result["raw"].usage_metadata["total_tokens"]

print(f"Content : {content}")
print(f"input_tokens : {input_tokens}")
print(f"output_tokens : {output_tokens}")
print(f"total_tokens : {total_tokens}")
```

```python
response = cloud_llm.chat.completions.parse(
    model="gpt-4o-mini",
    messages=messages,
    response_format=MacBookSpec,
)

content       = response.choices[0].message.parsed
input_tokens  = response.usage.prompt_tokens
output_tokens = response.usage.completion_tokens
total_tokens  = response.usage.total_tokens

print(f"Content : {content}")
print(f"input_tokens : {input_tokens}")
print(f"output_tokens : {output_tokens}")
print(f"total_tokens : {total_tokens}")
```
